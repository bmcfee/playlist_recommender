# MSD + AotM playlist features

This notebook implements MSD-AoTM data munging from the ground up using pandas dataframes.

```python
# Load in the tag data

def make_tag_frame(sqlitepath):
    with sqlite3.connect(sqlitepath) as dbc:
    
        # We need to add actual ids here, since the msd db files don't have keys
        tags = pd.read_sql('''SELECT tag FROM tags''', dbc)
        tags['tag_id'] = tags.index
    
        tracks = pd.read_sql('''SELECT tid as track_id FROM tids''', dbc)
        tracks['trackid_id'] = tracks.index
    
        # Offset by 1 because the sqlite database has 1-based counting
        track_to_tag = pd.read_sql('''SELECT (tid - 1) as trackid_id, (tag - 1) as tag_id, val as value FROM tid_tag''', dbc)
    
        # Merge the whole shebang
        full_tag_data = track_to_tag.merge(tags).merge(tracks)

        full_tag_data = full_tag_data[['tag', 'value', 'track_id']]
        
        
    return full_tag_data
```

```python
def reduce_tags(tag_data, k=10, min_score=None):
    
    if k is not None:
        reduced_tags = tag_data.sort(['track_id', 'value'], ascending=[True, False]).groupby('track_id').head(k)
    else:
        reduced_tags = tag_data
    
    if min_score is not None:
        reduced_tags = reduced_tags[reduced_tags.value >= min_score]
        
    return reduced_tags
```

```python
def min_tag_size(tag_data, min_size=1e3, key='tag'):
    
    big_enough = (tag_data.groupby([key]).size() >= min_size)
    
    return tag_data[tag_data[key].map(lambda x: big_enough[x])]
```

```python
import nltk

def longest_string(z):
    
    z = list(z)
    return z[np.argmax([len(_) for _ in z])]


def stem_tags(tag_data):
    
    tokenizer = nltk.tokenize.RegexpTokenizer('\w+')
    stemmer = nltk.stem.PorterStemmer()
    
    # 1: tokenize the tags
    # 2: stem the sub-tags
    # 3: recombine the sub-tags
    # 4: group the tag matrix
    #     aggregate(value) -> max
    #     aggregate(tag) -> argmax(len(x))

    tag_data['tag_stem'] = [''.join([stemmer.stem(_) for _ in tokenizer.tokenize(tag.lower())]) for tag in tag_data['tag']]

    #G = tag_data.groupby(by=['track_id', 'tag_stem'])
    #return G.aggregate({'value': np.max, 'tag': longest_string})
    expander = tag_data.groupby(by='tag_stem').aggregate({'tag': longest_string})
    #expander = expander.applymap(unicode.lower)
    
    #return expander
    tag_data['tag_stem'] = [expander.loc[_, 'tag'] for _ in list(tag_data['tag_stem'])]
    tag_data['tag_stem'] = tag_data['tag_stem'].apply(unicode.lower)
    
    G = tag_data.groupby(by=['track_id', 'tag_stem'], as_index=False)
    return G.aggregate({'value': np.max})
    #return tag_data
```

```python
def flag_mismatch(tracks, bad_matches_csv):
    
    bad_pairs = pd.DataFrame.from_csv(bad_matches_csv, header=None, index_col=None)
    
    bad_pairs.columns = ['song_id', 'track_id']
    bad_pairs = bad_pairs[['song_id']]
    bad_pairs['mismatch'] = True
    
    out = tracks.merge(bad_pairs, how='left', on=['song_id'])
    
    # Find all the multiply-matched tracks
    return out.merge(multi_songs(tracks), how='left').fillna(False)
```

```python
def multi_songs(tracks):
    
    tg = tracks.groupby('song_id').count()
    
    msd = pd.DataFrame(tg[tg['track_id'] > 1].index, columns=['song_id'])
    msd['multimatch'] = True
    return msd
```

```python
def make_tracks(unique_track_path):
    
    tracks = pd.DataFrame.from_csv(unique_track_path,
                                   sep='<SEP>',
                                   header=None,
                                   index_col=None)  

    tracks.columns = ['track_id', 'song_id', 'artist', 'title']
    
    return tracks
```

```python
def msd_metadata(dbfile):
    
    with sqlite3.connect(dbfile) as dbc:
        track_meta = pd.read_sql_query('SELECT track_id, song_id, duration, artist_familiarity, year FROM songs',
                                       dbc,
                                       index_col='track_id')

    track_meta.loc[track_meta['year'] == 0, 'year'] = pd.NaT
    return track_meta
```

```python
def make_playlist_segments(pl_data, tracks):
    
    bad_tracks = set(tracks[tracks.mismatch].song_id) | set(tracks[tracks.multimatch].song_id)
    
    segments = []
    
    for row_num, row_data in pl_data.iterrows():
        
        segment_num = 0
        for segment in row_data.filtered_lists:
            
            segment_position = 0
            for song_id in segment:
                if song_id in bad_tracks:
                    if segment_position > 0:
                        segment_num += 1
                        segment_position = 0
                else:
                    datum = {'mix_id': row_data.mix_id, 
                             'segment_id': segment_num,
                             'segment_position': segment_position,
                             'song_id': song_id}
                    segments.append(datum)
                    segment_position += 1
                
            segment_num += 1
            
    segments = pd.DataFrame.from_records(segments)
    
    return segments
```
