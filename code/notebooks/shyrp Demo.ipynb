```python
def make_counts(n, playlists):
    c = np.zeros(n)
    for pl in playlists:
        for idx in pl:
            c[idx] += 1
            
    return c
```

```python
dummy_playlists = {'user 1': [ [0, 1, 2, 3, 4],
                              [2, 3, 4],
                              [4, 9, 10] ],
                   
                   'user 2': [[7, 8, 9],
                              [6, 7, 8],
                              [9, 10, 11]],
                   'user 3': [[17,18,19],
                              [15,17]]}
```

```python
n_songs = 32
H = scipy.sparse.lil_matrix(np.zeros((n_songs,7)), shape=(n_songs,7), dtype=bool)
H[:, 0] = True
H[:3, 2] = True
H[8:13, 1] = True
H[::2, 3] = True
H[::3, 4] = True
H[::4, 5] = True
H[::5, 6] = True
H = H.tocsr()
```

```python
model = shyrp.PlaylistModel(params='ebus', 
                            n_factors=2,
                            verbose=logging.INFO,
                            max_iter=20)
```

```python
model.fit(dummy_playlists, H)
```

```python
k = 0
my_list = dummy_playlists['user 1'][k]
print model.loglikelihood(my_list, user_id='user 1', normalize=True)
print model.loglikelihood(my_list, user_id='user 2', normalize=True)
print model.loglikelihood(my_list, user_id='user 3', normalize=True)
print model.loglikelihood(my_list, normalize=True)
```

```python
plt.plot(np.exp(model.b_) / np.sum(np.exp(model.b_)), label='Model scores')
c = make_counts(len(model.b_), dummy_playlists['user 1'])
plt.plot(c / c.sum(), label='% Play counts user 1')
c = make_counts(len(model.b_), dummy_playlists['user 2'])
plt.plot(c / c.sum(), label='% Play counts user 2')
c = make_counts(len(model.b_), dummy_playlists['user 3'])
plt.plot(c / c.sum(), label='% Play counts user 3')
plt.legend()
```

```python
import gzip
import ujson as json
import collections
import cPickle as pickle
```

```python
def load_raw_aotm_data():
    with gzip.open('/home/bmcfee/git/bmcfee.github.io/data/aotm2011_playlists.json.gz', 'r') as f:
        data = json.load(f)
        
    return data
```

```python
def make_pl_data(J, max_users=None):
    
    playlists = collections.defaultdict(list)
    
    song_counter = 0
    
    songs = {}
    song_names = {}
    
    for i, pl in enumerate(J):
        if max_users and i >= max_users:
            break
            
        user_name = pl['user']['name']
        
        for l in pl['filtered_lists']:
            
            tracks = []
            for t in l:
                if t not in songs:
                    songs[t] = song_counter
                    
                    song_counter += 1
                    
                tracks.append(songs[t])
            playlists[user_name].append(tracks)
            
        
    return dict(playlists), songs
```

```python
def get_song_names(data):
    songs = {}
    
    for playlist in data:
        for info, song_id in playlist['playlist']:
            if song_id:
                songs[song_id] = u' - '.join(info)
    return songs
```

```python
J = load_raw_aotm_data()
```

```python
playlists, songs = make_pl_data(J, max_users=50)
```

```python
song_ids = dict([_[::-1] for _ in songs.items()])
song_names = get_song_names(J)
```
