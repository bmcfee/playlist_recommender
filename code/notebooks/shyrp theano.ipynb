# Okay, bad idea...

The first cut at the stochastic hypergraph random walk used a whole bag of tricks and doesn't seem to work that well.

What else to do?

# Theano

How about just doing gradient descent on the whole damn objective?

Ultimately, this is a deepish architecture with some non-linearities.  We should be able to do map optimization by any means necessary.

Recall that the total negative log-likelihood function factors into user-centric bigram sub-problems over triples (user_id, source_id, target_id).

The point-wise objective function looks like

$$
-\log P(Y = t ~|~ s, i) = -\log \sum_e P(Y = t ~|~ e, i) P(E = e ~|~ s)
$$

where the edge selection probability looks like

$$
P(E=e ~|~ s) = \frac{\exp(w_e) H_{s,e} }{\langle \exp(w),  H_{s,\cdot} \rangle} = \text{softmax}\left(H_{s, \cdot} \exp(w) \right)
$$

and the item selection probability looks like

$$
P(Y=t ~|~ e, i) = \frac{ \exp\left( U_i V_t^\mathsf{T} + b_t^\mathsf{T}\right) H_{t, e}}{
\langle \exp\left( U_i V^\mathsf{T} + b^\mathsf{T}\right),  H_{\cdot, e}\rangle
} = \text{softmax}\left(\exp\left(U_i V^\mathsf{T} + b^\mathsf{T}\right)H_{\cdot, e} \right)
$$

# Issues...

What to do with initial-state tracks?  Before, we could set `s=None` and not worry about it too much.

In Theano land, things are strongly typed, so this won't fly.  We'll need to hack around this somehow.  I'll use `s=-1` to indicate a start track, and hand-hack all-ones feasibility vectors.

# Re-writing the objective

$$
\log P(Y=t ~|~ s, i) = \log \sum_{e} P(Y=t ~|~ e, i) P(E=e ~|~ s)
$$

We can push the denominator of $P(Y=t~|~e,i)$ into the edge selection probability vector, since they have the same shape.  This simplifies the edge marginalization computation to a single dot product for each $(i,s,t)$ triple.

```python
n_users, n_items, n_factors, n_edges = 5, 32, 2, 10
```

```python
sigma_u, sigma_v, sigma_b, sigma_w = 1e-3, 1e-3, 1e-3, 1e-3
```

```python
H = scipy.sparse.lil_matrix(np.zeros((n_items,n_edges)), shape=(n_items,n_edges), dtype=np.int32)
H[:, 0] = True
H[:3, 2] = True
H[8:13, 1] = True
H[::2, 3] = True
H[::3, 4] = True
H[::4, 5] = True
H[::5, 6] = True
H = H.tocsr()
```

```python
def to_one_hot(y, nb_class, dtype=None):
    """Return a matrix where each row correspond to the one hot
    encoding of each element in y.

        :param y: A vector of integer value between 0 and nb_class - 1.
        :param nb_class: The number of class in y.
        :param dtype: The dtype of the returned matrix. Default floatX.

        :return: A matrix of shape (y.shape[0], nb_class), where each
          row ``i`` is the one hot encoding of the corresponding ``y[i]``
          value.

   """
    ret = theano.tensor.zeros((y.shape[0], nb_class),
                              dtype=dtype)
    ret = theano.tensor.set_subtensor(ret[theano.tensor.arange(y.shape[0]), y],
                                      1)
    return ret
```

```python
def sparse_slice_rows(H, idx):
    '''Returns a dense slice H[idx, :]'''
    
    vecs = to_one_hot(idx, H.shape[0], dtype=H.dtype)
    
    return ts.dot(vecs, H)
```

```python
n_factors = 4
```

```python
n_items, n_edges = H.shape

# Inputs
y_s = T.ivector(name='y_s')
y_t = T.ivector(name='y_t')
u_i = T.ivector(name='u_i')

# model parameters
U = theano.shared(np.zeros((n_users, n_factors)), name='U')
V = theano.shared(np.random.randn(n_items, n_factors), name='V')
b = theano.shared(np.zeros(n_items), name='b')
w = theano.shared(np.zeros(n_edges), name='w')

nll = []
```

```python
# Compute the intermediate distributions
rho = 1e-8

## item scores
item_scores = T.dot(U[u_i], V.T) + b

# stable exponentiation
e_scores = T.exp(item_scores - item_scores.max(axis=1, keepdims=True))

## edge feasibilities
# if y_s >= 0: H[y_s, :]
# if y_s < 0: ones(n_edges)

prev_feas = sparse_slice_rows(H, y_s)
prev_feas = theano.tensor.set_subtensor(prev_feas[y_s < 0, :], 1)

# raw edge probabilities
edge_given_prev = T.nnet.softmax(prev_feas * w)

# Compute edge normalization factors: 
#   sum of score mass in each edge for each user
edge_norms = ts.dot(e_scores, H)

# Next song feasibilities
next_feas = sparse_slice_rows(H, y_t)

# Slice the weight
next_weight = e_scores[T.arange(y_t.shape[0]), y_t]

# Marginalize
probs = next_weight * T.dot(next_feas, (edge_given_prev / (rho + edge_norms)).T)

ll = T.log(probs).mean()
```

```python
# Priors

u_prior = -0.5 * sigma_u * (U**2.0).sum()
v_prior = -0.5 * sigma_v * (V**2.0).sum()
b_prior = -0.5 * sigma_b * (b**2.0).sum()
w_prior = -0.5 * sigma_w * (w**2.0).sum()
```

```python
cost = -1 * (ll + u_prior + v_prior + b_prior + w_prior)
```

```python
#updates = nntools.updates.adadelta(-(ll + u_prior + v_prior + b_prior + w_prior),
#                                   [U, V, b, w])
#updates = nntools.updates.adagrad(cost, [U, V, b, w])
updates = nntools.updates.adagrad(cost, [U, V, b])
#updates = nntools.updates.adadelta(cost, [U, V, b])
#updates = nntools.updates.sgd(cost, [U, V, b, w], 1e-2)
```

```python
train = theano.function(inputs=[u_i, y_s, y_t], 
                        outputs=[ll], 
                        updates=updates)
```

```python
predict = theano.function(inputs=[u_i], outputs=T.argmax(item_scores, axis=1))
predict_unbiased = theano.function(inputs=[u_i], outputs=T.argmax(item_scores - b, axis=1))
predict_negbiased = theano.function(inputs=[u_i], outputs=T.argmax(item_scores - 2*b, axis=1))
```

```python
def playlist_to_bigram(playlist):
    '''Convert a sequence of ids into bigram form.

    A 'None' is pushed onto the front to indicate the beginning.
    '''

    my_pl = [-1]
    my_pl.extend(playlist)

    bigrams = zip(my_pl[:-1], my_pl[1:])

    return bigrams

def make_bigrams_usermap(playlists):
    '''generate user map and bigram lists.

    input:
        - playlists : dict : user_id => playlists

    '''

    user_map = dict()

    bigrams = []

    for i, user_id in enumerate(sorted(playlists)):

        user_map[user_id] = i

        user_bigrams = []
        for pl in playlists[user_id]:
            user_bigrams.extend(playlist_to_bigram(pl))

        bigrams.append(user_bigrams)

    return user_map, bigrams

def make_theano_inputs(playlists):
    
    usermap, bigrams = make_bigrams_usermap(playlists)
    
    users = []
    prevs = []
    nexts = []
    
    for user_id, bg in enumerate(bigrams):
        for s, t in bg:
            users.append(user_id)
            prevs.append(s)
            nexts.append(t)
            
    return (usermap,
            np.asarray(users, dtype=np.int32),
            np.asarray(prevs, dtype=np.int32),
            np.asarray(nexts, dtype=np.int32))
```

```python
def make_ix(users, nexts, n_users=None, n_items=None):
    
    if n_users is None:
        n_users = 1 + np.max(users)
        
    if n_items is None:
        n_items = 1 + np.max(nexts)
        
    X = np.zeros((n_users, n_items))
    
    for u_id, n in zip(users, nexts):
        X[u_id, n] += 1
        
    return X
```

```python
def load_raw_aotm_data():
    with gzip.open('/home/bmcfee/git/bmcfee.github.io/data/aotm2011_playlists.json.gz', 'r') as f:
        data = json.load(f)
        
    return data
```

```python
def make_pl_data(J, max_users=None):
    
    playlists = collections.defaultdict(list)
    
    song_counter = 0
    
    songs = {}
    song_names = {}
    
    for i, pl in enumerate(J):
        if max_users and i >= max_users:
            break
            
        user_name = pl['user']['name']
        
        for l in pl['filtered_lists']:
            
            tracks = []
            for t in l:
                if t not in songs:
                    songs[t] = song_counter
                    
                    song_counter += 1
                    
                tracks.append(songs[t])
            playlists[user_name].append(tracks)
            
        
    return dict(playlists), songs
```

```python
def get_most_popular(playlists, songs, n_users):
    
    # first pass: compute song popularities
    
    popularity = collections.defaultdict(int)
    
    for user, pls in playlists.iteritems():
        for pl in pls:
            for song in pl:
                popularity[song] += 1
                
    # Next, map users to their total track popularity
    
    user_pop = collections.defaultdict(int)
        
    for user, pls in playlists.iteritems():
        n_songs = 0.0
        for pl in pls:
            n_songs += len(pl)
            for song in pl:
                user_pop[user] += popularity[song]
                
        user_pop[user] /= n_songs
                
    kv = sorted([(v,k) for k,v in user_pop.iteritems()], reverse=True)
    
    top_users = [k for v,k in kv[:n_users]]
    
    top_playlists = {k: playlists[k] for k in top_users}
    inv_songs = {v: k for k, v in songs.items()}
    
    # Now remap the song ids
    newsongs = {}
    counter = 0
    for user, pls in top_playlists.iteritems():
        for pl in pls:
            for song in pl:
                if inv_songs[song] not in newsongs:
                    newsongs[inv_songs[song]] = counter
                    counter += 1
    
    new_playlists= {}
    for user, pls in top_playlists.iteritems():
        new_playlists[user] = []
        
        for pl in pls:
            new_playlists[user].append([newsongs[inv_songs[song]] for song in pl])
        
    return new_playlists, newsongs
```

```python
def get_song_names(data):
    songs = {}
    
    for playlist in data:
        for info, song_id in playlist['playlist']:
            if song_id:
                songs[song_id] = u' - '.join(info)
    return songs
```
