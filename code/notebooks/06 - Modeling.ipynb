```python
def decompose(df, song_map, max_users=np.inf):
    
    playlists = {}
    
    for count, user in enumerate(df.index.levels[0].unique()):
        
        if count >= max_users:
            break
            
        # Add the user to the playlist collection
        playlists.setdefault(user, [])
        
        # Slice the frame
        d_u = df.loc[user]
        
        for mix_id in d_u.index.get_level_values(0).unique():
            d_mix = d_u.loc[mix_id]
            
            for segment_id in d_mix.index.get_level_values(0).unique():
                playlists[user].append([song_map[_] for _ in d_mix.loc[segment_id]['song_id']])
                
    return playlists
```

```python
def graph_to_song_map(H):
    
    return dict([_[::-1] for _ in enumerate(H.index)])
    
```

```python
def load_edges(*files, **kwargs):
    
    min_size = kwargs.get('min_size', 318)
    
    frames = [pd.read_pickle(_) for _ in files]
    frames = [frame.drop(frame.columns[frame.count() < min_size], axis=1).to_dense()
              for frame in frames]
    
    H = frames.pop(0)
    
    while len(frames):
        H = H.join(frames.pop(0), how='outer', sort=True)
    
    H = H.to_sparse()
    return H
```

```python
songs = graph_to_song_map(H_frame)
```

```python
pl_train = pd.read_pickle('/home/bmcfee/git/playlist_recommender/data/folds/0_train.pickle')
```

```python
playlists = decompose(pl_train, songs, max_users=100)
```

```python
song_ids = dict([_[::-1] for _ in songs.items()])
```

```python
n_users = len(playlists)
```

```python
model_eb = shyrp.PlaylistModel(H, n_users, edge_reg=1e-6, bias_reg=1e-6,
                              n_factors=0,
                              n_epochs=int(10),
                              batch_size=128,
                              params='eb',
                              verbose=shyrp.logging.DEBUG)
```

```python
model_eb.fit(playlists)
```

```python
model_eb.loglikelihood(playlists)
```

```python
-np.log(H.shape[0])
```

```python
model_b = shyrp.PlaylistModel(H_u, n_users, bias_reg=1e-6,
                              n_factors=0,
                              n_epochs=int(10),
                              batch_size=128,
                              params='b',
                              verbose=shyrp.logging.DEBUG)
```

```python
model_b.fit(playlists)
```

```python
model_b.loglikelihood(playlists)
```

```python
mpld3.disable_notebook()
plt.hlines(-np.log(H.shape[0]), 0, len(model_eb.nll_), alpha=0.75, color='k', linestyle='--', label='Baseline')
plt.plot(model_eb.nll_, alpha=0.75, label='Bias + Edge')
plt.plot(model_b.nll_, alpha=0.75, color='g', label='Bias')
plt.legend(loc='best')
plt.ylabel('Training set log-likelihood')
plt.xlabel('Training batches')
plt.tight_layout()
```
