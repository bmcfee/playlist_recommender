# MSD + AotM playlist features

This notebook implements MSD-AoTM data munging from the ground up using pandas dataframes.

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn
import IPython.display
seaborn.set()
%matplotlib inline
np.set_printoptions(precision=4)

import cPickle as pickle
import gzip
import ujson as json
import sqlite3
```

```python
# Load in the tag data

def make_tag_frame(sqlitepath):
    with sqlite3.connect(sqlitepath) as dbc:
    
        # We need to add actual ids here, since the msd db files don't have keys
        tags = pd.read_sql('''SELECT tag FROM tags''', dbc)
        tags['tag_id'] = tags.index
    
        tracks = pd.read_sql('''SELECT tid as track_id FROM tids''', dbc)
        tracks['trackid_id'] = tracks.index
    
        # Offset by 1 because the sqlite database has 1-based counting
        track_to_tag = pd.read_sql('''SELECT (tid - 1) as trackid_id, (tag - 1) as tag_id, val as value FROM tid_tag''', dbc)
    
        # Merge the whole shebang
        full_tag_data = track_to_tag.merge(tags).merge(tracks)

        full_tag_data = full_tag_data[['tag', 'value', 'track_id']]
        
        
    return full_tag_data
```

```python
def reduce_tags(tag_data, k=10, min_score=None):
    
    if k is not None:
        reduced_tags = tag_data.sort(['track_id', 'value'], ascending=[True, False]).groupby('track_id').head(k)
    else:
        reduced_tags = tag_data
    
    if min_score is not None:
        reduced_tags = reduced_tags[reduced_tags.value >= min_score]
        
    return reduced_tags
```

```python
def min_tag_size(tag_data, min_size=1e3):
    
    big_enough = (tag_data.groupby(['tag']).size() >= min_size)
    
    return reduced_tags[tag_data.tag.map(lambda x: big_enough[x])]
```

```python
def flag_mismatch(tracks, bad_matches_csv):
    
    bad_pairs = pd.DataFrame.from_csv(bad_matches_csv, header=None, index_col=None)
    
    bad_pairs.columns = ['song_id', 'track_id']
    bad_pairs = bad_pairs[['song_id']]
    bad_pairs['mismatch'] = True
    
    out = tracks.merge(bad_pairs, how='left', on=['song_id'])
    
    out.mismatch = out.mismatch.fillna(False)
    
    return out
```

```python
def make_tracks(unique_track_path):
    
    tracks = pd.DataFrame.from_csv(unique_track_path,
                                   sep='<SEP>',
                                   header=None,
                                   index_col=None)  

    tracks.columns = ['track_id', 'song_id', 'artist', 'title']
    
    return tracks
```

```python
def make_playlist_segments(pl_data):
    
    segments = []
    
    for row_num, row_data in pl_data.iterrows():
        
        for segment_num, segment in enumerate(row_data.filtered_lists):
            
            for segment_position, song_id in enumerate(segment):
                datum = {'mix_id': row_data.mix_id, 
                         'segment_id': segment_num,
                         'segment_position': segment_position,
                         'song_id': song_id}
                segments.append(datum)
    segments = pd.DataFrame.from_records(segments)
    
    return segments
```

```python
with gzip.open('/home/bmcfee/git/bmcfee.github.io/data/aotm2011_playlists.json.gz', 'r') as f:
    pl_data = pd.DataFrame.from_records(json.load(f))
```

```python
pl_data.timestamp = pd.to_datetime(pl_data.timestamp)
```

```python
pl_data.user = pl_data.user.map(lambda x: x['name'])
```

```python
pl_data.category = pl_data.category.astype('category')
```

```python
pl_data = pl_data[['mix_id',
                   'user',
                   'category',
                   'filtered_lists',
                   'timestamp']].sort(columns=['mix_id']).reset_index(drop=True)
```

```python
pl_segments = pl_data.merge(segments)
del pl_segments['filtered_lists']
pl_segments['category'] = pl_segments['category'].astype('category')
```
